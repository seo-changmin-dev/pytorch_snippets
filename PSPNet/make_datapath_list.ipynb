```python
import os
import os.path as osp

import numpy as np

from PIL import Image
import matplotlib.pyplot as plt
```

```python
def make_datapath_list(root_path:str): # root_path = './data/VOCdevkit/VOC2012'
    # Segmentation 폴더에서 train.txt, val.txt 파일 불러오기
    train_txt_fpath = osp.join(root_path, 'ImageSets', 'Segmentation', 'train.txt')
    val_txt_fpath = osp.join(root_path, 'ImageSets', 'Segmentation', 'val.txt')

    train_img_list = list()
    train_anno_list = list()

    for data_id in open(train_txt_fpath):
        data_id = data_id.strip()
        
        # 그에 해당하는 이미지를 JPEGImages 폴더에서 파일 경로 찾기
        img_fpath = osp.join(root_path, 'JPEGImages', f"{data_id}.jpg")

        # 그에 해당하는 어노테이션을 SegmentationClass에서 파일 경로 찾기
        anno_fpath = osp.join(root_path, 'SegmentationClass', f"{data_id}.png")

        train_img_list.append(img_fpath)
        train_anno_list.append(anno_fpath)

    val_img_list = list()
    val_anno_list = list()

    for data_id in open(val_txt_fpath):
        data_id = data_id.strip()
        
        # 그에 해당하는 이미지를 JPEGImages 폴더에서 파일 경로 찾기
        img_fpath = osp.join(root_path, 'JPEGImages', f"{data_id}.jpg")

        # 그에 해당하는 어노테이션을 SegmentationClass에서 파일 경로 찾기
        anno_fpath = osp.join(root_path, 'SegmentationClass', f"{data_id}.png")

        val_img_list.append(img_fpath)
        val_anno_list.append(anno_fpath)

    # 리스트를 리턴하는 함수 만들기
    return train_img_list, train_anno_list, val_img_list, val_anno_list
    
    # 진짜 다 있는지 테스트 해보기 (코드에는 포함 X)
    # 잘 나오는지 테스트 해보기 (코드에는 포함 X)
```

```python
# 진짜 다 있는지 테스트 해보기 (코드에는 포함 X)
root_path = './data/VOCdevkit/VOC2012'

a, b, c, d = make_datapath_list(root_path)

for fpath_list in [a,b,c,d]:
    for fpath in fpath_list:
        if not osp.exists(fpath):
            print("Not Exist", fpath)

print("DONE")
```
